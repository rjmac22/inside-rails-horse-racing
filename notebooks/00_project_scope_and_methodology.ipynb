# 00 — Project Scope and Methodology

## Purpose

This notebook defines the scope, working principles and methodological rules for the Inside Rails horse-racing database project.

The project will convert a collection of third-party UK and Irish horse-racing data products into a documented, reproducible and professionally structured relational database.

This notebook does not clean or transform the racing data. Its purpose is to establish the rules that later notebooks must follow.


## Project Objective

The objective of Inside Rails is to build a professional relational database from historical UK and Irish horse-racing data.

The finished system should:

- preserve the original source files unchanged;
- document where every source file came from;
- make the structure and meaning of the supplied data understandable;
- separate raw, staging, core and analytical data;
- model racing entities and relationships explicitly;
- apply appropriate keys, constraints and indexes;
- record important data-quality issues without automatically “fixing” them;
- support reproducible analysis through Jupyter notebooks;
- demonstrate standard data-engineering and database-design techniques in a portfolio-ready form.

The immediate objective is not to produce betting predictions or racing strategies. Those may become later analytical uses of the database, but they must not determine how the foundational data model is built.


## Current Source Scope

The downloaded Kaggle dataset contains several related data products rather than one unified database.

The current primary source for the first phase of the project is:

`data/raw/form_2015-present/form_2015-present/raceform.db`

Other supplied products include:

- historical archives covering 1988–2004;
- historical archives covering 2005–2014;
- recent form HTML files;
- daily racecards;
- BHA ratings data;
- Betfair data.

These products will not be combined automatically.

Each source must first be treated as a separate product with its own:

- structure;
- date coverage;
- identifiers;
- update pattern;
- racing meaning;
- data-quality issues;
- potential overlaps with other products.

Relationships between products will only be created after their fields, entities and business meanings have been profiled and compared.


## Project Boundaries

### In scope

The project will:

- inspect and document the supplied source data;
- profile tables, columns, types, values and relationships;
- investigate racing-domain meaning before transformation;
- identify entities, candidate keys and business rules;
- design a relational target model;
- build reproducible staging, core and analytical layers;
- validate transformations with explicit tests;
- document limitations, assumptions and unresolved issues;
- preserve a clear audit trail from source to final database.

### Out of scope for the initial database build

The initial phase will not:

- build a betting model;
- claim predictive value from unvalidated fields;
- scrape additional racing websites;
- overwrite or manually edit raw source files;
- merge separate source products solely because their names appear similar;
- impute missing values without a defensible racing or analytical reason;
- remove duplicate-looking records before determining what they represent;
- treat unusual values as errors merely because they are statistical outliers.

The scope may be expanded later, but only through an explicit and documented decision.


## Methodological Principles

The project will follow these principles throughout the database build.

### 1. Preserve raw data

Files under `data/raw/` are treated as immutable source material. They will not be edited, renamed for convenience or overwritten by later processing.

### 2. Profile before cleaning

The structure, contents and apparent relationships of each source must be examined before cleaning rules are designed.

### 3. Understand meaning before transformation

A value that appears missing, duplicated or unusual may be valid in a horse-racing context. Transformations must therefore be based on both technical evidence and racing-domain meaning.

### 4. Separate observation from decision

Profiling notebooks should first show what is present in the data. Cleaning and modelling decisions should be recorded separately, with their reasoning and consequences made explicit.

### 5. Prefer reversible transformations

Early transformations should preserve source values and add standardised or interpreted fields alongside them. Destructive changes should be avoided unless the original value remains traceable.

### 6. Test important assumptions

Candidate keys, joins, date ranges, row counts, uniqueness rules and referential relationships must be tested rather than assumed.

### 7. Record uncertainty

Unresolved meanings, ambiguous identifiers and suspected data-quality problems should be documented as open issues instead of being hidden by premature corrections.

### 8. Keep the workflow reproducible

Each notebook should run from clearly defined inputs and produce repeatable outputs. Important logic should eventually be extracted into reusable code when it becomes stable.

### 9. Build for analysis without distorting the source

The final database should make racing analysis easier, but analytical convenience must not override source fidelity or sound relational design.


## Planned Data Architecture

The project will use a layered data architecture so that source preservation, cleaning, relational modelling and analysis remain distinct.

The architecture is conceptual at this stage. The final database technology has not yet been selected.

### Raw layer

The raw layer contains the original downloaded files exactly as supplied.

Purpose:

- preserve source evidence;
- support reproducibility;
- allow checks against later transformations;
- provide a recoverable starting point.

Raw files are not treated as the finished analytical database.

### Staging layer

The staging layer will reproduce source data in a controlled working environment.

It may add technical metadata such as:

- source file;
- source table;
- ingestion timestamp;
- source row identifier;
- source checksum or version;
- load status.

Source values should initially remain unchanged. Standardised fields may be added alongside them where useful.

### Core layer

The core layer will contain the professionally modelled relational database.

It will represent racing entities such as races, runners, horses, courses, trainers and jockeys only after the source evidence supports those distinctions.

The core layer should use:

- defined primary keys;
- documented business keys;
- foreign-key relationships;
- appropriate data types;
- constraints;
- indexes;
- controlled handling of missing and unknown values.

### Analytical layer

The analytical layer will provide convenient tables, views or derived datasets for research and reporting.

These may include:

- denormalised race-runner datasets;
- horse-form histories;
- trainer and jockey summaries;
- performance features;
- market and result comparisons.

Analytical outputs must remain traceable to the core and staging layers.

### Database technology decision

The supplied `raceform.db` file will initially be inspected as a SQLite source database.

This does not determine the technology used for the final staging, core or analytical layers.

The target platform will be selected later using explicit criteria, including:

- support for relational constraints and indexing;
- suitability for the data volume;
- reproducibility in Jupyter notebooks;
- portability;
- ease of testing and automation;
- analytical performance;
- portfolio relevance;
- deployment and maintenance requirements.

Possible options may include SQLite, DuckDB, PostgreSQL or another suitable relational platform. The choice will be documented as a separate architectural decision.

## Use of External Methodology

The project may draw on techniques from the 2024 book *Data Cleaning with Python for Data Analytics and Modeling* and from established data-engineering practice.

Ideas from external sources will be evaluated rather than copied automatically.

A technique will only be adopted when it:

- fits the structure and meaning of the racing data;
- preserves traceability to the original source;
- improves reproducibility or data quality;
- can be explained and tested;
- does not conceal uncertainty;
- remains appropriate for a relational database project.

Techniques will be adapted or rejected when they rely on assumptions that are unsuitable for this project.

In particular, the project will not assume that:

- missing values should always be imputed;
- duplicate-looking rows should always be deleted;
- outliers are necessarily errors;
- data should be altered merely to make a distribution more regular;
- every source column belongs in the final core model;
- generic cleaning rules are more reliable than racing-domain interpretation.

Where an external method influences a significant decision, the notebook should explain:

1. the original idea;
2. how it was adapted;
3. why it is appropriate here;
4. how the result was validated.


## Data-Quality Decision Framework

Potential data-quality problems will be classified before any corrective action is taken.

For each issue, the project should record:

- what was observed;
- which tables, columns and records are affected;
- whether the issue is technical, semantic or still uncertain;
- the likely cause;
- the possible effect on analysis;
- the proposed treatment;
- the evidence supporting that treatment;
- any information lost or changed by the treatment.

Possible outcomes include:

- retain the value unchanged;
- standardise the value while preserving the source version;
- flag the record for analytical caution;
- map the value to a documented controlled category;
- exclude the record from a specific analysis but retain it in the database;
- defer the decision until more racing-domain evidence is available;
- correct the value only when a defensible rule or authoritative reference exists.

No record should be deleted or altered solely because it looks inconvenient, rare or inconsistent.


## Keys, Identifiers and Relationships

Identifiers supplied by the source will be treated as evidence, not automatically accepted as reliable database keys.

For each candidate identifier, the project should test:

- whether it is populated;
- whether it is unique where uniqueness is expected;
- whether it remains stable across tables and source products;
- whether its meaning is documented or can be inferred safely;
- whether multiple identifiers represent the same real-world entity;
- whether the same identifier is reused in different contexts;
- whether joins based on it produce missing, duplicated or many-to-many matches.

The project will distinguish between:

- **source identifiers** — values supplied by the original data;
- **business keys** — meaningful combinations of fields that identify a racing event or entity;
- **surrogate keys** — database-generated identifiers used to support stable relational design.

Foreign-key relationships will only be enforced after their validity has been profiled and tested.

Names alone, such as horse, trainer, jockey or course names, will not be assumed to identify unique entities without supporting evidence.


## Missing Values, Duplicates and Outliers

Missing values, duplicate-looking records and unusual values will be investigated in context rather than handled by generic automatic rules.

### Missing values

A missing value may mean:

- the information was not recorded;
- the field was not applicable;
- the value was not yet known at the time of collection;
- the source used another field or code to represent the information;
- the record is incomplete;
- the source extraction failed.

These possibilities must be distinguished where practical. Missingness may itself contain useful information and should not be removed through automatic imputation.

### Duplicate-looking records

Rows that appear identical may represent:

- genuine duplicate ingestion;
- repeated snapshots of changing information;
- separate runners or races with incomplete identifiers;
- amended or republished results;
- records that differ only in fields not initially inspected.

Duplicates will therefore be tested against the expected grain of the table before any removal rule is considered.

### Outliers and unusual values

Extreme values may reflect:

- exceptional but valid race conditions or performances;
- different units or encodings;
- historical rule changes;
- data-entry or extraction errors;
- a field whose meaning has been misunderstood.

Statistical rarity alone is not sufficient evidence that a value is wrong.


## Validation and Reconciliation

Every major transformation should be accompanied by checks that show whether the result is complete, internally consistent and traceable to its source.

Validation should include, where appropriate:

- source and target row-count reconciliation;
- uniqueness tests for candidate and primary keys;
- null checks for required fields;
- accepted-value and range checks;
- referential-integrity checks;
- duplicate detection at the expected table grain;
- date and chronology checks;
- comparisons of totals before and after transformation;
- checks that excluded records are counted and explained;
- spot checks of representative and unusual records.

A successful notebook run is not, by itself, evidence that the data is correct.

Validation results should be visible in the notebook, and failed checks should stop or clearly invalidate the relevant stage rather than being silently ignored.


## Reproducibility and Project Structure

The project should be understandable and runnable by someone who did not create the original working environment.

To support this:

- notebooks should use project-relative paths rather than machine-specific absolute paths;
- environment and package requirements should be recorded;
- generated database layers, extracts and reports should be reproducible from documented inputs, regardless of the platform ultimately selected;
- raw source files should remain outside version control but be identifiable through checksums and inventories;
- reusable logic should move into Python modules once it is stable;
- exploratory code that no longer contributes to the final argument should be removed from completed notebooks;
- important parameters should be defined clearly near the start of each notebook;
- notebook outputs should be retained when they provide useful evidence;
- random processes, if introduced later, should use recorded seeds;
- changes to source scope, schema or methodology should be documented in version control.

The repository should contain the code, notebooks, metadata and documentation needed to explain the build, while excluding large or licensed source data that should not be redistributed.


## Open Architectural Decisions

The following decisions remain deliberately open at this stage:

- the target database platform;
- whether staging and core layers should use the same technology;
- whether analytical workloads should remain in the relational database or use a separate analytical engine;
- the final deployment model;
- the long-term update and ingestion process;
- how multiple historical and current source products should eventually be integrated.

These decisions will be made after the primary source database has been profiled and the likely data volume, structure, relationships and analytical requirements are better understood.

No technology should be selected merely because it is familiar or because the source happens to use it.


## Documentation and Decision Records

Important project decisions should be recorded close to the evidence that produced them.

Each significant decision should state:

- the issue being decided;
- the evidence considered;
- the chosen approach;
- alternatives rejected;
- assumptions or uncertainties;
- consequences for later stages;
- whether the decision should be revisited.

Examples include:

- choosing the grain of a core table;
- selecting a business or surrogate key;
- standardising a course or participant name;
- excluding a source field from the core model;
- separating two apparently similar source products;
- treating a missing value as unknown rather than not applicable;
- resolving overlapping historical records.

Minor implementation details may remain in notebook commentary or code comments. Decisions that affect the structure, interpretation or reproducibility of the database should be visible in markdown and version control.


## Definition of Done for a Project Stage

A notebook or database stage is complete only when:

- its objective is stated clearly;
- its input sources are identified;
- the relevant data has been profiled or transformed;
- important assumptions have been tested;
- validation results are visible;
- significant findings are interpreted;
- decisions and unresolved issues are documented;
- outputs are saved in their intended project location;
- the notebook runs from beginning to end without unexplained errors;
- the next project step is clear.

A stage is not complete merely because code has executed or a table has been created.

Where evidence is insufficient, the correct outcome may be a documented unresolved question rather than a forced conclusion.
